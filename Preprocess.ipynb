```python
def create_sub_df(dataframe, winning_bid_date, previous_date, author):
    sub_dataframe = dataframe.query(f"date >= '{previous_date}' and date <='{winning_bid_date}'")
    sub_dataframe['price'] = sub_dataframe['price'].astype('int')
    df_value = sub_dataframe['price'].sum()
    df_count = sub_dataframe['price'].count()
    
    sub_author_df = sub_dataframe.query(f"author=={author}")
    author_value = sub_author_df['price'].sum()
    author_count = sub_author_df['price'].count()
    
    return df_value, df_count, author_value, author_count

def subtract_days_from_date(date, days):
    """Subtract days from a date and return the date.
    
    Args: 
        date (string): Date string in YYYY-MM-DD format. 
        days (int): Number of days to subtract from date
    
    Returns: 
        date (date): Date in YYYY-MM-DD with X days subtracted. 
    """
    
    subtracted_date = pd.to_datetime(date) - timedelta(days=days)
    subtracted_date = subtracted_date.strftime("%Y-%m-%d")

    return subtracted_date

def previous(day_list, counter, year):
    previous_begin = ''
    previous_end = ''
    previous_year = ''
    if counter == 1:
        previous_begin = day_list[-1][0]
        previous_end = day_list[-1][1]
        previous_year = str(int(year)-1)
    else:
        previous_begin = day_list[counter-2][0]
        previous_end = day_list[counter-2][1]
        previous_year = str(year)
    return previous_begin, previous_end, previous_year
```

```python
def calc_avg_growth(merged_df, winning_bid_date, previous_date):
    sub_dataframe = merged_df.query(f"날짜 >= '{previous_date}' and 날짜 <'{winning_bid_date}'")
    avg = sub_dataframe['종가'].mean()
    now = fill_holiday_value(merged_df, winning_bid_date)
    day_before = fill_holiday_value(merged_df, previous_date)
    if day_before>0:
        growth = (now-day_before)/day_before
    else:
        growth = 0
    return growth, avg

def fill_holiday_value(df, date):
    value = 0.0
    day_counter = 1
    if len(merged_df[merged_df['날짜']==date])>0:
        value = merged_df[merged_df['날짜']==date]['종가'].values[0]
    else:
        while True:
            day_before = subtract_days_from_date(date, day_counter)
            if len(merged_df[merged_df['날짜']==day_before])>0:
                value = merged_df[merged_df['날짜']==day_before]['종가'].values[0]
                break
            else:
                day_counter+=1
    
    return value
```

```python
def relative_strength_index(df, merged_df, day_param, name):
    # data: data set (pandas dataframe을 가정)
    # day_param: RSI를 구하고자 하는 기간
    # use_col: RSI을 구하고자 하는 열을 담은 list 

    #가격이 전일 가격보다 상승한 날의 상승분은 U(up) 값이라고 하고,
    #가격이 전일 가격보다 하락한 날의 하락분은 D(down) 값이라고 한다.
    #day_param은 기간을 며칠 동안으로 할 것인가가 있다. Welles Wilder는 14일을 사용할 것을 권유했다. 대체로 사용되는 값은 9일, 14~15일, 25~28일 등이다.   
    merged_df['U'] = ''
    merged_df['D'] = ''
    for j in range(1, len(merged_df)):
        if merged_df['종가'][j]-merged_df['종가'][j-1]>0:
            merged_df['U'][j] = merged_df['종가'][j]-merged_df['종가'][j-1]
            merged_df['D'][j] = 0
        else:
            merged_df['U'][j] = 0
            merged_df['D'][j] = merged_df['종가'][j-1]-merged_df['종가'][j]
    # U값과 D값의 평균값을 구하여 그것을 각각 AU(average ups)와 AD(average downs)라 한다.
    # AU를 AD값으로 나눈 것을 RS(relative strength) 값이라고 한다. RS 값이 크다는 것은 일정 기간 하락한 폭보다 상승한 폭이 크다는 것을 의미한다.
    # 다음 계산에 의하여 RSI 값을 구한다. RSI = RS / (1 + RS)
    merged_df[f'AU_{day_param}'] = ''
    merged_df[f'AD_{day_param}'] = ''
    merged_df[f'RS_{day_param}'] = ''
    merged_df[f'RSI_{day_param}_{name}'] = ''
    for j in range(day_param, len(merged_df)):
        merged_df[f'AU_{day_param}'][j] = sum(merged_df['U'][j:j+day_param])/day_param
        merged_df[f'AD_{day_param}'][j] = sum(merged_df['D'][j:j+day_param])/day_param
        merged_df[f'RS_{day_param}'][j] = merged_df[f'AU_{day_param}'][j]/merged_df[f'AD_{day_param}'][j]
        merged_df[f'RSI_{day_param}_{name}'][j] = merged_df[f'RS_{day_param}'][j]/(1+merged_df[f'RS_{day_param}'][j])

    
    df = pd.merge(df,merged_df, how='left', left_on='date', right_on='날짜')
    df.drop(['U','D'])
    return df

def bollinger_band(df, merged_df, name):
    # 볼린저 밴드의 상한선: 20일 이평선 값 + ( 20일 동안의 주가 표준편차 값 ) * 2
    # 볼린저 밴드의 하한선: 20일 이평선 값 - ( 20일 동안의 주가 표준편차 값 ) * 2
    # perb = %b = (주가 – 하한선) / (상한선 – 하한선) = (close - lbb) / (ubb - lbb)
        # 1. 하한선에 있는 상태에는 0
        # 2. 상한선에 있는 상태에는 1
        # 3. 중심에 있는 상태에는 0.5
        # 4. 하한선을 벗어난 상태에는 0 이하 (음수)
        # 5. 상한선을 벗어난 상태에는 1이상
    # bw = 밴드폭 (Bandwidth) = (상한선 – 하한선) / 중심선 = (ubb - lbb) / mbb
    # 밴드폭 (Bandwidth)은 상한선과 하한선의 밴드폭이 얼마나 큰지를 나타내 주는 지표이다.
    # 주가의 변동폭이 클 경우 볼린저 밴드의 폭이 넓어지고,
    # 주가의 변동폭이 적을 때는 볼린저 밴드의 폭 또한 좁아진다.    
    day_param = 20
    merged_df[f'bollinger_upper_limit_{name}'] = ''
    merged_df[f'bollinger_lower_limit_{name}'] = ''
    merged_df[f'bollinger_perb_{name}'] = ''
    merged_df[f'bollinger_bandwidth_{name}'] = ''
    
    for i in range(day_param):
        ma20 = sum(merged_df['종가'][i-20:i])
        merged_df[f'bollinger_upper_limit_{name}'][i] = ma20+np.std(merged_df['종가'][i-20:i])*2
        merged_df[f'bollinger_lower_limit_{name}'][i] = ma20-np.std(merged_df['종가'][i-20:i])*2
        merged_df[f'bollinger_perb_{name}'][i] = (merged_df['종가'][i]-merged_df[f'bollinger_lower_limit_{name}'][i])/(merged_df[f'bollinger_upper_limit_{name}'][i]-merged_df[f'bollinger_lower_limit_{name}'][i])
        merged_df[f'bollinger_bandwidth_{name}'][i] = (merged_df[f'bollinger_upper_limit_{name}'][i]-merged_df[f'bollinger_lower_limit_{name}'][i])/ma20
        
    df = pd.merge(df,merged_df, how='left', left_on='date', right_on='날짜')

    return df
```
